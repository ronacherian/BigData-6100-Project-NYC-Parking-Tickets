```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df2016 = pd.read_csv("Training dataset.csv")
```

```python
d = {'Unique Entry': df2016.nunique(axis = 0),
        'Nan Entry': df2016.isnull().any()}
pd.DataFrame(data = d, index = df2016.columns.values)
```

```python
x_ticks = df2016['Registration State_fix'].value_counts().index
heights = df2016['Registration State_fix'].value_counts()
y_pos = np.arange(len(x_ticks))
fig = plt.figure(figsize=(15,14)) 
# Create horizontal bars
plt.barh(y_pos, heights)
 
# Create names on the y-axis
plt.yticks(y_pos, x_ticks)
 
# Show graphic
plt.show()
```

```python
pd.DataFrame(df2016['Registration State_fix'].value_counts()/len(df2016)).nlargest(10, columns = ['Registration State_fix'])
```

```python
x_ticks = df2016['Month of the Year'].value_counts().index
heights = df2016['Month of the Year'].value_counts()
y_pos = np.arange(len(x_ticks))
fig = plt.figure(figsize=(15,14)) 
# Create horizontal bars
plt.barh(y_pos, heights)
 
# Create names on the y-axis
plt.yticks(y_pos, x_ticks)
 
# Show graphic
plt.show()
```
